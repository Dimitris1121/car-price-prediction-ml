```python
# Determining categorical features
nums = ['int64', 'float64']
categorical_columns = []
features = df.columns.values.tolist()
for col in features:
    if df[col].dtype in nums: continue
    categorical_columns.append(col)


categorical_columns
```

```python
# Transforming categorical features
# labelEncoding
for col in categorical_columns:
    if col in df.columns:
        from sklearn import preprocessing
        le = preprocessing.LabelEncoder()
        le.fit(list(df[col].astype(str).values))
        df[col] = le.transform(list(df[col].astype(str).values))

df.head()
```

```python
df.info()
```

```python
# Transforming the rest variables to ints
df['odometer'] = df['odometer'].astype(int)
df['year'] = df['year'].astype(int)

df.head(5)
```

```python
# Preparation for Models
# Importing libraries

from sklearn.model_selection import train_test_split


from sklearn.linear_model import LinearRegression

from sklearn.ensemble import RandomForestRegressor

from sklearn.tree import DecisionTreeRegressor

import lightgbm as ltb


from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn import metrics
```

```python
# Seperating Target Variable from the rest.
# First attempt with all variables.
y= df.price
X= df.drop('price',axis=1)
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
```

```python
print('Linear Regression (using all variables)')

LR_1_R2 = lin_reg.score(X_test,y_test)*100
print('R Squared:', LR_1_R2)

LR_1_MAE = round(metrics.mean_absolute_error(y_test, y_pred),2)
print('Mean Absolute Error:', LR_1_MAE)

LR_1_RMSE = round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2)
print('Root Mean Squared Error:', LR_1_RMSE)
```

```python
#Random Forrest
rf_reg = RandomForestRegressor(n_estimators=20, random_state=0)
rf_reg.fit(X_train, y_train)
y_pred = rf_reg.predict(X_test)
```

```python
print('Random Forrest (using all variables)')

RF_1_R2 = rf_reg.score(X_test,y_test)*100
print('R Squared:', RF_1_R2)

RF_1_MAE = round(metrics.mean_absolute_error(y_test, y_pred),2)
print('Mean Absolute Error:', RF_1_MAE)

RF_1_RMSE = round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2)
print('Root Mean Squared Error:', RF_1_RMSE)
```

```python
# Decision Tree
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
y_pred = dt_reg.predict(X_test)
```

```python
print('Decision Tree (using all variables)')

DT_1_R2 = dt_reg.score(X_test,y_test)*100
print('R Squared:', DT_1_R2)

DT_1_MAE = round(metrics.mean_absolute_error(y_test, y_pred),2)
print('Mean Absolute Error:', DT_1_MAE)

DT_1_RMSE = round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2)
print('Root Mean Squared Error:', DT_1_RMSE)
```

```python
# LightGBM
lgbm_reg  = ltb.LGBMRegressor()
lgbm_reg.fit(X_train, y_train)
y_pred = lgbm_reg.predict(X_test)
```

```python
print('LightGBM (using all variables)')

LGBM_1_R2 = lgbm_reg.score(X_test,y_test)*100
print('R Squared:', DT_1_R2)

LGBM_1_MAE = round(metrics.mean_absolute_error(y_test, y_pred),2)
print('Mean Absolute Error:', DT_1_MAE)

LGBM_1_RMSE = round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2)
print('Root Mean Squared Error:', DT_1_RMSE)
```
